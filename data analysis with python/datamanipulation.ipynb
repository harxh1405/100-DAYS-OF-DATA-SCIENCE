```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('ecommerce.csv')
df.head()
```

```python
df.describe() ###statistical summary of numerical columns
```

```python
df.dtypes
```

```python
##handling missing values
df.isnull().any(axis=1) ##rowwise
```

```python
df.isnull().any()##columnwise
```

```python
df.isnull().sum()
```

```python
df_filled = df.fillna(0) ###filling missing values with 0
df_filled.head()
```

```python
###filling missing values with mean of the column
df['total_spent_fillna'] = df['total_spent'].fillna(df['total_spent'].mean())
df.head()
```

```python
## renaming columns
df_renamed = df.rename(columns={'total_spent':'total_amount_spent'})
df_renamed.head()
```

```python
##changing data types
df['age']= df['age'].fillna(df['age'].mean()).astype(int)

df
```

```python
df['new_value'] = df['age'].apply(lambda x: x*2)
df.head()
```

```python
###data aggregating and grouping
df.groupby('country')['total_spent'].mean()
```

```python
df.groupby(['country','device_type'])['total_spent'].sum()
```

```python
###aggregate multiple functions
grouped_agg = df.groupby('country')['total_spent'].agg(['mean', 'sum', 'count'])
grouped_agg
```

```python
##merging and joining dataframes
df1 = pd.DataFrame({
    'customer_id':[1,2,3,4],
    'customer_name':['Alice', 'Bob', 'Charlie', 'David']
})
df2 = pd.DataFrame({
    'customer_id':[1,2,3,5],
    'total_spent':[250, 150, 300, 400]
})
```

```python
##merging on customer_id columns
pd.merge(df1, df2, on='customer_id', how='inner')
```

```python
pd.merge(df1, df2, on='customer_id', how='outer') ###outer join
```

```python
pd.merge(df1, df2, on='customer_id', how='left')
```

```python
pd.merge(df1, df2, on='customer_id', how='right')
```
